# Entrada da regress√£o linear

```python
import math
import matplotlib.pyplot as plt

dataset = [
    (1, 2),
    (2, 3),
    (4.5, 3.5),
    (5.5, 4),
    (6.2, 7.5),
    (7.8, 9.7)
]
dataset
```

```python
alfa = 0.01

erro_anterior = 0
```

```python
x1l = []
x2l = []

for linha in dataset:
    x1l.append(linha[0])
    x2l.append(linha[1])

plt.scatter(x1l, x2l, c='blue')
```

# Elabora√ß√£o do treinamento do algoritmo!

```python
epoca = 0
m = 0
b = 0

xr = np.linspace(0, 9, 10)

while epoca < 10000:
    # montando as colunas 
    e   = []
    ex1 = []
    e2  = []
    for linha in dataset:
        x1 = linha[0]
        x2 = linha[1]
        # calcula o valor da hip√≥tese atual
        x2_ = m * x1 + b
        # erro da inst√£ncia atual
        ei = x2 - x2_
        e.append(ei)
        ex1.append(ei * x1)
        e2.append(ei ** 2)
        
    emq = sum(e2) / len(e2)
        
    if math.fabs(emq - erro_anterior) < 0.001:
            break
    
    epoca += 1
    erro_anterior = emq
    x2r = m * xr + b
    
    plt.scatter(x1l, x2l, c='blue')
    plt.plot(xr, x2r, c='green')
    plt.show()
    # atualizar o m e o b
    m = m + alfa*sum(ex1)
    b = b + alfa*sum(e)
    print(epoca, ":", emq) 
    
    
        
```

# Fase de testes 
Posso prever valores que n√£o estavam no dataset!

```python
testes = [2, 3, 5.6, 7.8, 2.4, 6.7, 9.0 , 30]

for teste in testes:
    print(teste, ":", teste*m + b)
```
