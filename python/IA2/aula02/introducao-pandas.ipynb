# Estrutura de um dataframe:

* Existe um dicionário, contendo como chave as colunas.
* E para cada chave, existe uma lista que representa os dados da coluna.

### CUIDADO! AS COLUNAS DEVEM TER MESMO TAMANHO

```python
df = pd.DataFrame({
    "x1" : [1, 2, 3, 4, 5, 6, 7, 8],
    "x2" : [2, 4, 6, 8, 10, 12, 14, 16],
    "x3" : [1, 3, 5, 7, 8, 9, 3, 2],
    "x4" : ["A", "B", "C", "D", "E", "F", "G", "H"]
})
df
```

# Obter colunas do dataframe

```python
df["x2"]
```

```python
classe = df["x4"]
classe
```

```python
df[["x1", "x3", "x4"]]
```

# Seleção de linhas pode ser feita pelo valor do índice

```python
df[0:6]
```

```python
df[0:3]
```

# Função "loc"/"iloc" para seleção de linhas e colunas ... 

Seleção de uma linha específica

```python
df.loc[3]
```

Seleção de uma intervalo de linhas

```python
df.loc[3:5]
```

```python
df.loc[0:5:2]
```

```python
df.loc[0:5:2, "x1"]
```

```python
df.loc[[0, 3, 6]]
```

```python
df.loc[[0, 4], ["x2", "x4"]]
```

```python
df.iloc[:, 0:3]
```

```python
df.loc[:, ["x1", "x4"]] # equivalente df[["x1", "x4"]]
```

# Consultas em um dataframe

```python
selecionados = [True, True, True, False, False, True, True, False]
df[selecionados]
```

```python
df["x1"] > 5
```

```python
df[df["x1"] > 5]
```

```python
df[(df["x1"] > 5) & (df["x3"] < 5)]
```

```python
df[(df["x2"] == 4) | (df["x3"] > 7)]
```

```python
df[~(df["x1"] == 4)]
```
